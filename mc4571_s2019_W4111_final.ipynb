# Query Question I

## Question

- The career summary integrates information from Lahman2017. The relevant tables are:
    - People
    - Appearances
    - Batting
    - Pitching
    - Fielding
    
    
- A sample of running the query is below.


- Columns:
    - _Games_ is $G\_all$ from _Appeareances._
    - Batting: _Hits_ is sum of _H,_ _At Bats_ is sum of _AB,_ _RBIs_ is sum of _RBI._ _Average_ is the sum of _H_ divided by the sum of _AB._
    - Pitching: $P\_wins$ is the sum of _W_ and $P\_loss$ is the sum of _L._
    - Fielding: _Put Outs_ is sum of _PO,_ _Assists_ is sum of _A_ and _Errors_ is some of _E._
    
    
- All numeric types must be integers, except for _Average._ You must round _Average_ to 3 decimal places.


- Your column names must match.


- You should be able to figure out what "--" means and how to produce it.

## Answer

````python
%%sql

CREATE VIEW `lahman2017`.`player_performance_summary` AS
    SELECT 
		`e`.`playerID` as `playerID`,
        `e`.`teamID` as `teamID`,
        `e`.`yearID` as `yearID`,
        `e`.`Games` as `Games`,
        `e`.`Hits` as `Hits`,
        `e`.`At Bats` as `At Bats`,
        `e`.`HRs` as `HRs`,
        `e`.`RBIs` as `RBIs`,
        `e`.`P_wins` as `P_wins`,
        `e`.`P_loss` as `P_loss`,
        `f`.`Put Outs` as `Put Outs`,
        `f`.`Assists` as `Assists`,
        `f`.`Errors` as `Errors`
	FROM
		(((SELECT 
			`c`.`playerID` as `playerID`,
			`c`.`teamID` as `teamID`,
			`c`.`yearID` as `yearID`,
			`c`.`Games` as `Games`,
			`c`.`Hits` as `Hits`,
			`c`.`At Bats` as `At Bats`,
			`c`.`HRs` as `HRs`,
			`c`.`RBIs` as `RBIs`,
			`d`.`P_wins` as `P_wins`,
			`d`.`P_loss` as `P_loss`
		FROM
			(((SELECT
				`a`.`playerID` as `playerID`,
				`a`.`teamID` as `teamID`,
				`a`.`yearID` as `yearID`,
				`a`.`Games` as `Games`,
				`b`.`Hits` as `Hits`,
				`b`.`At Bats` as `At Bats`,
				`b`.`HRs` as `HRs`,
				`b`.`RBIs` as `RBIs`
		FROM
			(((SELECT
			`lahman2017`.`appearances`.`playerID` as `playerID`,
				`lahman2017`.`appearances`.`teamID` as `teamID`,
                `lahman2017`.`appearances`.`yearID` as `yearID`,
                `lahman2017`.`appearances`.`G_all` as `Games`
		FROM
			`lahman2017`.`appearances`))`a`
		LEFT JOIN(SELECT
			`lahman2017`.`batting`.`playerID` AS `playerID`,
            `lahman2017`.`batting`.`teamID` AS `teamID`,
            `lahman2017`.`batting`.`yearID` AS `yearID`,
            CAST(SUM(`lahman2017`.`batting`.`H`) as UNSIGNED) AS `Hits`,
            CAST(SUM(`lahman2017`.`batting`.`AB`) as UNSIGNED) AS `At Bats`,
            CAST(SUM(`lahman2017`.`batting`.`HR`) as UNSIGNED) AS `HRs`,
			CAST(SUM(`lahman2017`.`batting`.`RBI`) as UNSIGNED) AS `RBIs`
		FROM
			`lahman2017`.`batting`
		GROUP BY `lahman2017`.`batting`.`playerID`, `lahman2017`.`batting`.`teamID`, `lahman2017`.`batting`.`yearID`)`b` ON (((`a`.`playerID` = `b`.`playerID`)
			AND (`a`.`teamID` = `b`.`teamID`)
            AND (`a`.`yearID` = `b`.`yearID`))))))`c`
		LEFT JOIN (SELECT
			`lahman2017`.`pitching`.`playerID` as `playerid`,
				`lahman2017`.`pitching`.`teamID` as `teamid`,
                `lahman2017`.`pitching`.`yearID` as `yearid`,
                SUM(`lahman2017`.`pitching`.`W`) as `P_Wins`,
				SUM(`lahman2017`.`pitching`.`L`) as `P_loss`
		FROM 
			`lahman2017`.`pitching`
		GROUP BY `lahman2017`.`pitching`.`playerID`, `lahman2017`.`pitching`.`teamID`, `lahman2017`.`pitching`.`yearID`) `d` ON (((`c`.`playerID` = `d`.`playerID`)
			AND (`c`.`teamID` = `d`.`teamid`)
            AND (`c`.`yearID` = `d`.`yearid`))))))`e`
		LEFT JOIN (SELECT
			`lahman2017`.`fielding`.`playerID` as `playerid`,
				`lahman2017`.`fielding`.`teamID` as `teamid`,
                `lahman2017`.`fielding`.`yearID` as `yearid`,
                CAST(SUM(`lahman2017`.`fielding`.`PO`) as UNSIGNED) as `Put Outs`,
				CAST(SUM(`lahman2017`.`fielding`.`A`) as UNSIGNED) as `Assists`,
				CAST(SUM(`lahman2017`.`fielding`.`E`) as UNSIGNED) as `Errors`
		FROM 
			`lahman2017`.`fielding`
		GROUP BY `lahman2017`.`fielding`.`playerID`, `lahman2017`.`fielding`.`teamID`, `lahman2017`.`fielding`.`yearID`) `f` ON (((`e`.`playerID` = `f`.`playerID`)
			AND (`e`.`teamID` = `f`.`teamid`)
            AND (`e`.`yearID` = `f`.`yearid`))))
````

````python

%%sql
CREATE VIEW `lahman2017`.`player_summary_query` AS
    SELECT 
        (SELECT `lahman2017`.`people`.`nameLast`
            FROM 
                `lahman2017`.`people`
            WHERE
                (`lahman2017`.`people`.`playerID` = `player_performance_summary`.`playerID`)) as `last_name`,
        (SELECT 
            `lahman2017`.`people`.`nameFirst`
            FROM
                `lahman2017`.`people`
            WHERE
                (`lahman2017`.`people`.`playerID` = `player_performance_summary`.`playerID`)) as `first_name`,
            `player_performance_summary`.`playerID` as `playerID`,
            `player_performance_summary`.`teamID` as `teamID`,
            `player_performance_summary`.`yearID` as `yearID`,
            `player_performance_summary`.`Games` as `Games`,
            `player_performance_summary`.`Hits` as `Hits`,
            `player_performance_summary`.`At Bats` as `At Bats`,
            `player_performance_summary`.`Hits` / `player_performance_summary`.`At Bats` as `average`,
            `player_performance_summary`.`HRs` as `HRs`,
            `player_performance_summary`.`RBIs` as `RBIs`,
            `player_performance_summary`.`P_wins` as `P_wins`,
            `player_performance_summary`.`P_loss` as `P_loss`,
            `player_performance_summary`.`Put Outs` as `Put Outs`,
            `player_performance_summary`.`Assists` as `Assists`,
            `player_performance_summary`.`Errors` as `Errors`

        FROM 
            `lahman2017`.`player_performance_summary`
````

# Query Question II

## Question

- In the simple version of Moneyball, we tried to predict a team's wins based on a team's performance statistics.


- A simple sample of the initial data we loaded is:

````python
%%sql

select teamid, yearid, w, ab, h, `2b` as b2, `3b` as b3, BB, hr, era
    from lahman2017.teams
    where yearid >= 2000
    order by yearid asc
    limit 10;
````

- The Boston Red Sox hit 167 homeruns (hr) in 2000. Was 167 homeruns good or bad? You cannot answer the question in isolation. The answer depends on the homerun production of other teams.


- Normalizing performance require two steps:
    1. Computing average and aggregate behavior in a season over all teams.
    2. Understanding how a team's metrics compared to aggregates and averages for the season.
    
    
- Write __a single query__ that produces a relative performance table per team, per year.
    - For w, ab, h, b2, b3, hr, and bb, the corresponding column is the team's value divided by the total over all teams for the year.
    - For ERA, the value is the team's ERA relative to the average ERA over all teams for the year.

## Answer

````python
%%sql

select
    teamid, yearid,
    round(t.w/averages.w,3) as w,
    round(t.ab/averages.ab, 3) as ab,
    round(t.h/averages.h, 3) as h,
    round(t.b2/averages.b2, 3) as b2,
    round(t.b3/averages.b3, 3) as b3,
    round(t.hr/averages.hr, 3) as hr,
    round(t.bb/averages.bb, 3) as bb,
    round(t.era/averages.era, 3) as rel_era from
    (SELECT
        yearid, sum(w) as w, sum(ab) as ab,
         sum(h) as h, sum(`2b`) as b2, sum(`3b`) as b3, sum(BB) as BB,
             sum(hr) as hr, round(avg(era), 3) as era
        from lahman2017.teams
        where yearid >= 2000
        group by yearid) as averages
join
    (select teamid, yearid, w, ab, h, `2b` as b2, `3b` as b3, BB, hr, era
        from lahman2017.teams
        where yearid >= 2000) as t
using (yearid)

limit 20;
````

## Answers

#### 1) Give three benefits and three disadvantages to normalizing a database schema

Normalizing database schema provides many benefits. 
- 1) The most notable is to remove redundancies and inconsistencies within the data.
- 2) Normalizing data schema also provides better clarity to understand how the objects-to-data are mapped. These results promote loose coupling between classes and provide more easily understandable data.
- 3) Because the database normalizing breaks down larger tables to smaller tables, the user/DBA who maintaing these databases is afforded even more flexibility in terms of modifying these structures. From a logistical perspective, it is much easier to modify a small table with little data than to modify one big table that holds all of the dat. 


Three disadvantages:
- 1) A setback of having highly normalized data schema is slower reporting/accessing performance because the data is stored in different rows and tables, compared to a single row in a single table. When normalized data is stored in many different rows and tables, updates to the database schema are difficult to maintain.
- 2) In order to merge these smaller tables, we must do table joins. This may result in overly-complicated, tedious queries which may also increase read times and as a DBA/user who is writing these joins, we are prone to make more error as well. 
- 3) Since joins are required to merge various smaller tables, indexing does not work as efficiently. From an efficiency perspective, read times slower because the joins dont typically work well with indexing.

#### 2) Explain the concept of _functional dependencies_ and how they relate to the concept of _keys._

A functional dependency simply means that for every unique value of B exists for every unique value of A. Keys are essentially inputs, labelled as X, which can be labelled as superkeys, if and only if, it determines all other attributes of R (RHS of any functional dependency).

#### 1)

Since we are dealing with a large collection of rows, it is not beneficial for us to simply repeatedly insert rows. Doing so is an extremely inefficient and slow process. What we can do is actually perform a process called bulk loading where we sort the rows by primary key first, which allows us to sequentially store leaves and also avoid having to do periodic reorganization of rows in order for other rows to be inserted later on (which what we would have to do if the rows were not sorted by PK first). 

#### 2)

Linear probing and quadratic probing are both methods that allow us to address collisions when it comes to hashing. Lienar probing seeks the next possible free slot in the case that a collision occurs. Whereas, for quadratic probing, we will look for the i^2 th slot in the i'th iteration. For instance, if there is a collision at hash(x) then we will check hash(x) + 1^2 next. The advantage of quadratic probing over linear probing is that we can often avoid clustering, which is when maany consecutive elements forms groups and the time that it takes to find the next free slot or search for an element increases substantially.

#### 3)
 
The 'Steal/No Force' buffer force management policy consists of No Force and Steal, which allows for the highest performance. No force ensures that in the case that a system crashes before it the modifications are written to disk, as little as possible is written in a convenient place. Whereas, steal happens when the working memory is full and then the T1 does so by kicking off some other processes off the working memory to a more stable storage. 
 
The benefits of using this method for crash recovery is that we get a speed boost by not commmitting everything to memory. The downside is that we risk writing possibly uncommitted data to memeory. 

The DMBS will often create checkpoints so that we can better recover in the event of system failures. Furthermore, DBMS will often follow ACID-compliancy that guarentees that all of a transaction succeeds or none does, that the data is consistent, transactions are independent of one another, and that once a transaction is committed it will remain in the system even if there is a system failure right after.

#### 5) Briefly explain a a Grace Hash Join. What is the main benefit and how does it achieve it?

A grace hash join allows us to hash partition both relations on the join attribute using the same hash function. The result is that each relation ends up being hashed into k partitions, which are then written to disk. The main benefit is that grace hash join avoids having to rescan the entire S relation by first partitioning both R, S via the same hash function and then sending all these sub-buckets (of both the hashed R, S) to the disk. Then the last step is load the pairs of buckets and build a new hash table for the smaller partitioned relation, and then follow up with probing the other relation for matches with the other hash table. The end result is that R tuples in partition i will only match S tuples in partition i due to how the buckets were formed (based on hashing on the join key). This second hash is what provides extensive in-memory benefits.

#### 1) What is the difference between Two-Phase Locking (2PL) and Strict Two-Phase Locking? What condition to Strict 2PL prevent that 2PL does not prevent?

For the two-phase locking, the locking protocal divides the execution phase of the transaction into three parts. During the first part, the transactions seeks permission for the locks it requires and is growing (acquisition phase). Afterwards in the second part, the transaction will have acquired all its locks. Then when the transaction releases its first lock, it cannot request another one, and the rest of the locks held by the transaction are released and during this phase, shrinking occurs. 

Meanwhile during the strict two-phase locking, the first phase is the same as the 2PL. However, after acquiring all the locks in the acquisition phase, Strict 2PL does not release a lock after using it. In fact, it actually will hold all the locks until it commits to releases them all at once. 

The condition that Strict 2PL prevents that 2PL does not prevent is the cascading abort. When I have released the lock but have not committed the data, then we have unlocked an updated piece of data so that someone else can actually read that uncommitted piece of data. In this case, if I abort then they have to abort. This is cascading abort.  

#### 2) What are deadlocks? What are two techniques for detecting and resolving deadlocks?


Deadlocks occur when transactions cannot make progress due to locks in such a way that neither transaction can move on. For instance, if we have two transactions, T1 and T2 where v = R(A) and v2 = R(B). If T1 tries to write to B, it cannot since B has a read lock on it and there is now a deadlock as T1 waits for T2 to release the lock. Most deadlocks occur between pairs of transactions. 

Two techniques for detecting and resolving deadlocks:
- Timeout: If we request a lock and we wait longer than the timeout then the database will abort the transaction. For instance, if in our previous example, T1 waits longer than x time designated by the timeout, then T2 would get aborted and the read block on B is released and then it can write to B. 
- Wait for graph: In our previous example, T1 is waiting for a lock that T2 has' and lets just say T2 then needs to write to A and there is a write lock on A. Here we can construct a wait-for graph that has two nodes, T1 and T2 and arrows pointing from T1->T2 and T2->T1. If the wait-for graph ever has a cycle (which in this case it does), then there is a deadlock present since all entities in the cycle are waiting on anoter transaction that is waiting to release the lock and the cycle continues. Resolve this by aborting the transaction.

A precedence graph has a node for each committed transaction in S and an arc from Ti to Tj if an action of Ti precedes and conflicts with one of of Tj's actions and it basically detects conflicts between transactions (ie. if one transactions is supposed to read a piece of data and the other transaction is supposed to write it).

A nonconflicting operation is one where two different transactions do not operate on the same data item. For instance, W1(A) and W2(B) or W1(A) and r2(B) is non conflicting, whereas R1(A) and W2(A) are conflicting. Property: If the conflict graph does not have a cycle, then it is conflict serializable.

#### Three examples of OLAP queries

OLAP queries ask in dimensional space what facts you want to find, based on where it appears int he dimensions and what aggegate operators (sum, min, max, average). Blelow are three examples:

Sum dollar value of products sold to 
    - customer in france 
    - by a store in france
    - in 1st Quarter of June 2017
    - product major_code = movie
    

Average unit_price of products sold to 
    - customers in Japan
    - from a tea shop in Tokyo, Japan
    - in December 2016
    - product major_code = tea-sets

Average quantity (number of units) of products sold to 
    - customers in Manhattan, NY 
    - from a pet store in Manhattan, NY
    - in second half of June 2014
    - supplier = West Paw Design
    - product major_code = pet toy
    - product minor_code = chew bones
    


#### Rollup:
Rollup allows us to aggregate many categories into one. For instance, if in our OLAP query we are dealing with customers from many different customers, a roll up basically says that instead of having a separate entry for customers from France, customers from Japan, customers from Korea etc, we can opt to collapse all these countries into an aggregate entry (which can then be named whatever we choose, ie. FJK for France, Japan, Korea).


#### Drill down
Drill down is the opposite of rollup. If we know the facts about FJK for instance, we can actually expand this so we can see the facts about the individual countries/categories.

#### Slice and Dice
These are techniques that alllows userse to essentially take out a slice of a set of data so that you can essentially view the data from various perspectives. For instance, in our second OLAP query, if we decided to simply have 'Average unit_price of products sold to customers in Japan' with only that one dimension (customers in japan), we are effectively slicing it in that one direction and fixing this aspect.

Meanwhile, dice is an operation that allows us to analyze the data from multiple dimensions and basically combine a bunch of slices together. For dice, we can pick specific values for multiple dimensions, similar to what we did above for the three queries.
